```python
import sqlite3
import matplotlib.pyplot as plt

# Connessione al database
conn = sqlite3.connect('temperature_data_1node.db')
cursor = conn.cursor()

# Esempio di query per selezionare i dati desiderati
cursor.execute("SELECT node_name, id_attuale, future_value, temperatura_float FROM temperatures")
rows = cursor.fetchall()

# Chiudi la connessione
conn.close()

# Dizionario per salvare i dati per ogni node_name
data_per_node = {}

# Organizza i dati per node_name
for row in rows:
    node_name, id_attuale, future_value, temperatura_float = row
    if node_name not in data_per_node:
        data_per_node[node_name] = {
            'ids': [],
            'temperature_attuale': [],
            'temperature_future': []
        }
    data_per_node[node_name]['ids'].append(id_attuale)
    data_per_node[node_name]['temperature_attuale'].append(temperatura_float)
    data_per_node[node_name]['temperature_future'].append(future_value)

# Creazione dei grafici sovrapposti per ogni node_name
for node_name, data in data_per_node.items():
    ids = data['ids']
    temperature_attuale = data['temperature_attuale']
    temperature_future = data['temperature_future']

    # Creazione del grafico sovrapposto
    plt.figure(figsize=(10, 6))

    # Curva per le temperature attuali
    plt.plot(ids, temperature_attuale, marker='o', linestyle='-', color='b', label='Temperatura Attuale')

    # Curva per i valori futuri
    plt.plot(ids, temperature_future, marker='o', linestyle='-', color='g', label='Future Value')

    plt.title(f"Grafico delle Temperature per {node_name}")
    plt.xlabel('ID')
    plt.ylabel('Temperatura')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
